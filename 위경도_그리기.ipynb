```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib
matplotlib.rcParams['font.family'].insert(0, 'Malgun Gothic')
%matplotlib inline
import folium
from pyproj import Proj, transform
# https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=vieo90&logNo=221320796491
```

```python
map1 = folium.Map(location=[37.566535, 126.977969199999936], zoom_start=15)
# map1
```

```python
def convert_5179(x, y):
  inProj = Proj(init='epsg:5179')
  outProj = Proj(init='epsg:4326')
  x2, y2 = transform(inProj, outProj, x, y)
  return [y2, x2]
```

```python
def convert_5186(x, y):
  inProj = Proj(init='epsg:5186')
  outProj = Proj(init='epsg:4326')
  x2, y2 = transform(inProj, outProj, x, y)
  return [y2, x2]
```

```python
df['위경도'] = df.apply(lambda loc: convert_5179(loc['X_COORD(X좌표)'], loc['Y_COORD(Y좌표)']), axis=1)
```

```python
KT_df['위경도'] = KT_df.apply(lambda loc: convert_5186(loc['X(X좌표)'], loc['Y(Y좌표)']), axis=1)
```

```python
# df['위경도'].head()
# folium으로 위경도 표시   
idx = 0
for i in df.iterrows():
  if idx == 5:
    break
  folium.Marker(i[1]['위경도'], popup=str(i[1]['STD_YMD(년월일)'])).add_to(map1)
  idx += 1
map1
```

```python
# df['위경도'].head()
# folium으로 위경도 표시   
idx = 0
for i in KT_df.iterrows():
  if idx == 10:
    break
  folium.Marker(i[1]['위경도'], popup=str(i[1]['ID(셀ID)'])).add_to(map1)
  idx += 1
map1
```
