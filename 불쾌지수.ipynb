```python
hum_percent = []
for avg, min in zip(humidity['평균'], humidity['최소']):
    _per_hum = min / avg * 100
    hum_percent.append(_per_hum)
hum_percent
```

```python
discom = 0
discomfort = []

for tem, hum in zip(temp['온도'], humidity['평균']):
    print(f'tem {tem} hum {hum}')
    discom = 0.81 * tem + 0.01*hum * (0.99*tem - 14.3) + 46.3
    discomfort.append(discom)
discomfort
```

```python
date = []
for i in temp['날짜']:
    date.append(str(i)[:7])
```

```python
data = pd.DataFrame({
    'date': date,
    'discomfort': discomfort
})
```
