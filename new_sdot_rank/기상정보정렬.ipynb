```python
def sort_temp_ILL_UV(filename):
    s_dot_loc = pd.read_csv('S-DoT-locationm.csv', encoding= 'euc-kr')
    df = pd.read_csv(filename, encoding='euc-kr')
    df.rename(columns = {'시리얼':'시리얼번호'}, inplace = True)
    df.drop(['기관 명', '송신 서버 번호','데이터 번호', '모델명', '구분', '초미세먼지(㎍/㎥)', '미세먼지(㎍/㎥)', '상대습도( %)', '풍향(°)', '풍속(m/s)', '돌풍 풍향(°)', 
               '돌풍 풍속(m/s)', '소음(dB)', '진동_x(g)', '진동_y(g)', '진동_z(g)', '진동_x 최대(g)', '진동_y 최대(g)', '진동_z 최대(g)', '흑구 운도(℃)', '초미세먼지 보정(㎍/㎥)', '미세먼지 보정(㎍/㎥)', '전송시간', '등록일자'], axis=1, inplace=True)
    df['조도(lux)'] = df['조도(lux)'].replace(np.nan, 0)
    df['자외선(UVI)'] = df['자외선(UVI)'].replace(np.nan, 0)
    s_dotgroup = df.groupby('시리얼번호').agg({
        '기온(℃)': 'mean',
        '조도(lux)': 'mean',
        '자외선(UVI)': 'mean'
        })
    s_dotgroup_big = s_dotgroup.sort_values(by ="자외선(UVI)",ascending = False)
    
    s_dot_loc.drop(['Unnamed: 5','Unnamed: 6',
       'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11',
       'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15',
       'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19',
       'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22'], axis='columns', inplace=True)
    
    # 같은것 모음
    s_dot_location_group = s_dot_loc.groupby('시리얼번호').sum()
    s_dot_complete = s_dotgroup_big.merge(s_dot_loc, left_on='시리얼번호',right_on='시리얼번호')

    return s_dot_complete
```

```python
path2 = 'S-DoT_NATURE_2021.06.28-07.04.csv'
df2 = pd.read_csv(path2, encoding='euc-kr')
preprocessing = sort_temp_ILL_UV(path2)
sorted_df = preprocessing.sort_values(by='시리얼번호')
sorted_df = sorted_df.replace(0, np.nan)
sorted_df = sorted_df.dropna(axis=0)
sorted_df.head()
```

```python
serial = sorted_df['시리얼번호'].values[:5]
temp = sorted_df['기온(℃)'].values[:5]
lux = sorted_df['조도(lux)'].values[:5]
uv = sorted_df['자외선(UVI)'].values[:5]

new_df = pd.DataFrame({
    '시리얼번호': serial,
    '기온': temp,
    '조도': lux,
    '자외선': uv
})
new_df.head()
```

```python
length = np.arange(1, 6)
temp_sort = new_df.sort_values(by=['기온'])
temp_sort['기온 순위'] = length

lux_sort = temp_sort.sort_values(by=['조도'])
lux_sort['조도 순위'] = length

uv_sort = lux_sort.sort_values(by=['자외선'])
uv_sort['자외선 순위'] = length

uv_sort['기온X가중치'] = uv_sort['기온 순위'] * 0.551
uv_sort['조도X가중치'] = uv_sort['조도 순위'] * 0.209
uv_sort['자외선X가중치'] = uv_sort['자외선 순위'] * 0.240
uv_sort['가중치 합'] = uv_sort['기온X가중치'] + uv_sort['조도X가중치'] + uv_sort['자외선X가중치']
sort_final = uv_sort.sort_values(by='가중치 합')
sort_final
```
